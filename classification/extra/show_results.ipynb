```python
generic_models = [
    "bert-base-uncased",
    "bert-base-cased",
    "FacebookAI/roberta-base",
    "FacebookAI/xlm-roberta-base",
    "FacebookAI/roberta-large",
    "FacebookAI/xlm-roberta-large",
]

sci_models = [
    "allenai/scibert_scivocab_cased",
    "allenai/scibert_scivocab_uncased",
]

cti_models = [
    "s2w-ai/DarkBERT",
    "jackaduma/SecBERT",
    "jackaduma/SecRoBERTa",
    "markusbayer/CySecBERT",
    "local/CyBERT-Base-MLM-v1.1",
    "local/tram_finetuned",
    "ehsanaghaei/SecureBERT"
]

def get_type(name):
    if name in cti_models:
        return "CTI"
    elif name in sci_models:
        return "Sci"
    else:
        return "Generic"

all_models_zero_weight = tram_swipe[(tram_swipe['pos_weight'] == 0) & (tram_swipe['model'] != "local/tram_finetuned")]
all_models_zero_weight['type'] = all_models_zero_weight['model'].apply(lambda x: get_type(x))
all_models_zero_weight['type'] = all_models_zero_weight['model'].apply(lambda x: get_type(x))
all_models_zero_weight['type'] = all_models_zero_weight['model'].apply(lambda x: get_type(x))
type_colors = {'CTI': 'blue', 'Sci': 'orange', 'Generic': 'gray'}
all_models_zero_weight['color'] = all_models_zero_weight['type'].map(type_colors)
```

```python
import matplotlib.pyplot as plt
df = all_models_zero_weight[all_models_zero_weight['dataset'] == "tram"]

# Plot
plt.figure(figsize=(4, 4))
plt.barh(df['model'], df['f1_mean'], color=df['color'], edgecolor='black', zorder=3)

# Add labels and title
plt.xlabel('F1-Score')
plt.ylabel('Model')
# plt.title('F1-Score of supervised fine-tuned models on TRAM dataset')

handles = [plt.Line2D([0], [0], color=color, lw=4) for color in type_colors.values()]
labels = type_colors.keys()

# plot show grid
plt.grid(axis='x', color='gray', linestyle='--', zorder=0)

plt.legend(handles, labels, title="Type", loc='upper left')
plt.savefig("plots/sft_bosch.pdf", bbox_inches='tight')
plt.show()
```

```python
generic_models = [
    "sentence-transformers/all-mpnet-base-v2",
    "nvidia/NV-Embed-V2"
]

cti_models = [
    "basel/ATTACK-BERT",
    "qcri-cs/SentSecBert_10k",
]

def get_type(name):
    if name in cti_models:
        return "CTI"
    elif name in sci_models:
        return "Sci"
    else:
        return "Generic"

sentence_similarity['type'] = sentence_similarity['model_name'].apply(lambda x: get_type(x))
sentence_similarity['type'] = sentence_similarity['model_name'].apply(lambda x: get_type(x))
sentence_similarity['type'] = sentence_similarity['model_name'].apply(lambda x: get_type(x))
type_colors = {'CTI': 'blue', 'Sci': 'orange', 'Generic': 'gray'}
sentence_similarity['color'] = sentence_similarity['type'].map(type_colors)
```

```python
import matplotlib.pyplot as plt
df = sentence_similarity[(sentence_similarity['dataset_name'] == "tram") & (sentence_similarity['label_set'] == "tram")]

# Plot
plt.figure(figsize=(4, 4))
plt.barh(df['model_name'], df['f1'], color=df['color'], edgecolor='black', zorder=3)

# Add labels and title
plt.xlabel('F1-Score')
plt.ylabel('Model')

handles = [plt.Line2D([0], [0], color=color, lw=4) for color in type_colors.values()]
labels = type_colors.keys()

# plot show grid
plt.grid(axis='x', color='gray', linestyle='--', zorder=0)
plt.legend(handles, labels, title="Type", loc='upper left')
plt.savefig("plots/ss_bosch.pdf", bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame for the first set of models
df1 = all_models_zero_weight

# Sample DataFrame for the second set of models
df2 = sentence_similarity

# Sample DataFrame for the first set of models
# df3 = all_models_with_weight

# Predefined mapping for label_set to number of labels
label_map = {
    'bosch_t': 118,
    'bosch_t10': 10,
    'bosch_t25': 25,
    'bosch_t50': 50,
    'bosch+tram': 143,
    'all_mitre': 637
}

# Map `label_set` to the number of labels for both datasets
df1['num_labels'] = df1['dataset'].map(label_map)
# df3['num_labels'] = df3['dataset'].map(label_map)
df2['num_labels'] = df2['label_set'].map(label_map)

# Calculate avg, min, and max f1 scores grouped by number of labels for both datasets
grouped1 = df1.groupby('num_labels')['f1_mean'].agg(['mean', 'min', 'max']).reset_index()
grouped1.rename(columns={'mean': 'avg_f1', 'min': 'min_f1', 'max': 'max_f1'}, inplace=True)

grouped2 = df2.groupby('num_labels')['f1'].agg(['mean', 'min', 'max']).reset_index()
grouped2.rename(columns={'mean': 'avg_f1', 'min': 'min_f1', 'max': 'max_f1'}, inplace=True)

# grouped3 = df3.groupby('num_labels')['f1_mean'].agg(['mean', 'min', 'max']).reset_index()
# grouped3.rename(columns={'mean': 'avg_f1', 'min': 'min_f1', 'max': 'max_f1'}, inplace=True)


# Plot
plt.figure(figsize=(10, 6))

# First dataset
plt.plot(grouped1['num_labels'], grouped1['avg_f1'], marker='o', color="blue", label='Supervised fine-tuning', linestyle='--')
plt.errorbar(
    grouped1['num_labels'], grouped1['avg_f1'],
    yerr=[grouped1['avg_f1'] - grouped1['min_f1'], grouped1['max_f1'] - grouped1['avg_f1']],
    fmt='o', capsize=5, color="blue"
)

# Second dataset
plt.plot(grouped2['num_labels'], grouped2['avg_f1'], marker='s', label='Sentence similarity', linestyle='--', color="red")
plt.errorbar(
    grouped2['num_labels'], grouped2['avg_f1'],
    yerr=[grouped2['avg_f1'] - grouped2['min_f1'], grouped2['max_f1'] - grouped2['avg_f1']],
    fmt='s', capsize=5, color="red"
)

# Second dataset
# plt.plot(grouped3['num_labels'], grouped3['avg_f1'], marker='^', label='Supervised fine-tuning (pos_weight 20)', linestyle='--', color="purple")
# plt.errorbar(
#     grouped3['num_labels'], grouped3['avg_f1'],
#     yerr=[grouped3['avg_f1'] - grouped3['min_f1'], grouped3['max_f1'] - grouped3['avg_f1']],
#     fmt='^', capsize=5, color="purple"
# )

# Add labels, title, and legend
plt.xlabel('Number of Labels')
plt.xticks([10, 25, 50, 118, 143, 637])
plt.ylabel('Average F1 Score')
plt.title('F1 Score vs. Number of Labels for Two Sets of Models')
plt.legend()
plt.grid(True)
plt.savefig("plots/ss_vs_sft_per_label.png", bbox_inches='tight')
plt.show()
```
